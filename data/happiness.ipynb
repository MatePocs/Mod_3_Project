```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://en.wikipedia.org/wiki/World_Happiness_Report"
```

```python
website_url = requests.get(url).text
soup = BeautifulSoup(website_url,'lxml')
table = soup.find("table",{"class":"wikitable sortable"})
body = table.find("tbody")
rows = body.findChildren(['th', 'tr'])
```

```python
countries = []
happiness_list = []
GDP_per_capita_list = []
social_support_list = []
healthy_life_exp_list = []
freedom_list = []
generosity_list = []
corruption_list = []
```

```python
for row in rows:
    cells = row.findChildren(['td'])
    if len(cells) > 0:
        
        country = cells[1].findAll('a')[0].text
        if country == '':
            country = cells[1].findAll('a')[1].text
        countries.append(country)
        
        happiness = cells[2].text.split('\n')[0]
        happiness_list.append(happiness)
        
        GDP_per_capita = cells[3].text.split('\n')[0]
        GDP_per_capita_list.append(GDP_per_capita)
        
        social_support = cells[4].text.split('\n')[0]
        social_support_list.append(social_support)
        
        healthy_life_exp = cells[5].text.split('\n')[0]
        healthy_life_exp_list.append(healthy_life_exp)
        
        freedom = cells[6].text.split('\n')[0]
        freedom_list.append(freedom)
        
        generosity = cells[7].text.split('\n')[0]
        generosity_list.append(generosity)
        
        corruption = cells[8].text.split('\n')[0]
        corruption_list.append(corruption)
```

```python
df = pd.DataFrame()
df['countries'] = countries
df['happiness'] = happiness_list
df['GDP_per_capita'] = GDP_per_capita_list
df['social_support'] = social_support_list
df['healthy_life_exp'] = healthy_life_exp_list
df['freedom'] = freedom_list
df['generosity'] = generosity_list
df['corruption'] = corruption_list
```
