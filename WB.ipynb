# API exploration

```python
wb.get_indicators().head()
```

```python
countries = wb.get_countries().head()
```

```python
countries.to_csv('./data/Countries.csv')
```

## Internet access

```python
internet = wb.search('access to internet')
internet
```

```python
df = wb.download(indicator=['gwp2_n','gwp2_n_1','gwp2_n_2','gwp2_n_3','gwp2_n_4','gwp2_n_5','gwp2_n_6','gwp2_n_7','gwp2_n'],country='all',start=2017,end=2017)
df.columns = [title for title in internet['name']]
df = df.dropna()
df
```

```python
df.to_csv('./data/Access_to_internet.csv')
```

## GDP (cross-referencing with Wikipedia)

```python
gdp = wb.search('GDP per capita')
```

```python
df = wb.download(indicator=['NY.GDP.PCAP.CD'],country='all',start=2018,end=2019)
df.columns = gdp[gdp['id']=='NY.GDP.PCAP.CD']['name']
df = df.dropna()
df
```

```python
df.to_csv('./data/GDP.csv')
```
